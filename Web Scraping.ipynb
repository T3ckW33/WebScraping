```python
import numpy as np
import pandas as pd
import requests
from bs4 import BeautifulSoup
import re
```

```python
#baseurl= 'https://www.senheng.com.my/'

headers = {
    "User-Agent": 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/88.0.4324.190 Safari/537.36'
}



productlinks = []  

for x in range (1,10):
    r = requests.get(f'https://www.senheng.com.my/all-products/home-entertainment.html?p={x}')
    soup = BeautifulSoup(r.content, 'lxml')

    productlist = soup.find_all('li', class_='item product product-item')
    
 

    for item in productlist:
        for link in item.find_all('a',class_='product photo product-item-photo',href=True):
            productlinks.append(link['href'])
            
print(len(productlinks))


#testlink= 'https://www.senheng.com.my/samsung-50-inch-tu8500-crystal-uhd-4k-smart-tv-2020.html'
```

```python
Productlist = []
for link in productlinks:

    r = requests.get(link, headers= headers)

    soup = BeautifulSoup(r.content, 'lxml')

    name = soup.find('h1',class_='page-title').text.strip()

    try:
        available = soup.find('div',class_='stock available').text.strip()
    except: available = "Out of Stock"
    
    try:
        special_price = soup.find('span',class_='special-price').text.strip()
    except: special_price = "0"
        
    try:
        original_price = soup.find('span',class_='old-price').text.strip()
    except: original_price = "0"
        
    try:
        selling_price = soup.find('span',class_='price-container price-final_price tax weee',itemprop="offers").text.strip()
    except: selling_price = "0"
    
    try:
        rating = soup.find('div',class_='rating-result').text.strip()
    except: rating = "0"
        
    try:
        review_number = soup.find('span',itemprop='reviewCount').text.strip()
    except: review_number = "0"

    try:    
        warranty = soup.find('div',class_='value',itemprop='warranty').text.strip()
    except: warranty = "0"
        
    try:
        category = soup.find('li',class_="item 3").text.strip()
    except: category = ""
        
    try:
        brand = soup.find('td', class_='col data').text.strip()
    except:
        brand = 'Not Categorized'
        

    Product = {'name' : name,
          'available' : available,
          'special_price' : special_price,
          'original_price' : original_price,
          'selling_price' : selling_price,
          'rating' : rating,
          'review_number' : review_number,
          'warranty' : warranty,
          'category' : category,'brand' : brand
         }

    Productlist.append(Product)

df = pd.DataFrame(Productlist)
print(df)
```

```python
df.to_csv('dataset_completed.csv', index=False)
```

```python
data = pd.read_csv('C:/Users/lowteckwee/Documents/Senheng_Data.csv')
```

```python
data.dtypes
```

```python
data['special_price'] = data['special_price'].str.replace('[^\d.]+','')
data['special_price'] = pd.to_numeric(data['special_price'], errors='coerce')#errors='coerce' = make empty value become NA
data['special_price'] = data['special_price'].fillna(0)
```

```python
data['original_price'] = data['original_price'].str.replace('[^\d.]+','')
data['original_price'] = pd.to_numeric(data['original_price'], errors='coerce')
data['original_price'] = data['original_price'].fillna(0)
```

```python
data['selling_price'] = data['selling_price'].str.replace("[^\d.]+", '')
data['selling_price'] = pd.to_numeric(data['selling_price'], errors='coerce')
```

```python
data.loc[data.name == "Philips 65 Inch 4K UHD LED Smart TV 65PUT6654/68","brand"]="Philips"
data.loc[data.name == "Philips 55 inch 4K UHD LED Smart TV 55PUT6654/68","brand"]="Philips"
data.loc[data.name == "Panasonic 40-Inch G300 LED TV TH-40G300K","brand"]="Panasonic"
data.loc[data.name == 'Panasonic 43" HX655 4K HDR Android TV TH-43HX655K â€“ Google Assistant & Chromecast',"brand"]='Panasonic'
data['brand'] = data['brand'].str.replace('96.8',"Panasonic")
data["brand"] = data['brand'].str.replace("-- Please Select --", 'Unclassified')

data['category'].fillna("uncategorized", inplace = True)
```

```python
data['warranty']=data['warranty'].str.extract('(\d+)',expand=False)
data['warranty'].fillna("0", inplace = True)
data['warranty'] = data['warranty'].astype('int')


data['rating'] = data['rating'].str.replace('% of 100','')
data['rating'] = data['rating'].str.replace('Excellent\n\n5','100')
data['rating'] = pd.to_numeric(data['rating'], errors='coerce')


data.loc[data['special_price'] == 0.00, 'original_price'] = 0
data.loc[data['original_price'] == 0.00, 'special_price'] = 0


Discount_percent = (data['original_price'] - data['special_price'])/data['original_price']
data['Discount_percent'] = Discount_percent
data['Discount_percent'] = pd.to_numeric(data['Discount_percent'], errors='coerce')
data['Discount_percent'] = data['Discount_percent'].fillna(0)


#data['available'] = data['available'].str.replace('not available','Out of stock')
#data= data.rename(columns={'available': 'availability'})

print(data)
```

```python
data.to_csv('Processed_Senheng_Data.csv', index=False)
```
